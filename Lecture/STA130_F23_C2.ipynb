## SQL: structured query language
- `select` [columns] (`.loc[:,]` which supports "boolean indexing" unlike `.iloc[]`)
- `where` [rows] (`[]` or `.loc[:,]` but not `.iloc[]`]
- `groupby` then "aggregation function"

```python
amazonbooks.select('Author')
```

```python
# tuple (1,2,3) # immutable
# list [1,2,3] # mutable
amazonbooks.loc[amazonbooks.isnull().sum(axis=1)>0, ('Author', 'Title')]
```

### What we learned

- boolean indexing
- `.iloc[]` selecting based on index
- `.loc[]` selecting with boolean indexing or columns or both
- `[]` selecting with boolean indexing or columns but not both
- `.groupby()` and "aggregation functions" like ~~`.count()`~~ `.size()`, `.mean()`, and `.std()`
- adding/creating columns
- learned about accessing columns
    - `df.col` or `df['col']` or `df[('col')]` or `df[(('col'))]` give series
    - `df[['col']]` gives dataframe
- `sort_values(ascending=True)`
