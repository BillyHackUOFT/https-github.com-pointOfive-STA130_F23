```python
#SAMPLE SOLUTION
np.random.seed(seed=10)
reps = 1000
sample_proportions = []
for i in range(reps):
    sample = np.random.choice(['Twitterer', 'Non_Twitterer'], size=400, replace=True, p=[prob, 1-prob])
    sample_prop = np.count_nonzero(sample == 'Twitterer') / 400
    sample_proportions.append(sample_prop)

dataframe = pd.DataFrame({'Proportion of Twitter Users': sample_proportions})
fig = px.histogram(dataframe, x='Proportion of Twitter Users', nbins=30, histnorm = 'probability density')
fig.update_layout(title='$\\text{Sampling distribution of } \hat{p} \\text{ assuming } H_0 \\text{ is true: the woman is guessing}$', 
                  xaxis_title='$\\hat{p} \\text{ values for 12 "50/50" chance guesses}$', yaxis_title='Density')
fig.show()
```

```python
# SAMPLE SOLUTION
test_stat = 11
expected_under_null = 6
sum(abs(sim_guess_correct[i] - expected_under_null) >= abs(test_stat-expected_under_null) for i in range(reps)) / reps
```

SAMPLE SOLUTION  
(a) p-value of 0.0053   
(b) Yes, can use the same simulation data  
(c) Still very unlikely, with only about 5 out of 1000 (e.g., our simulated sampling distribution shows 46 out of 10,000) people would be able to guess at least 11/12 correctly (or only 1 or 0 correctly). Strong evidence against the null hypothesis that the woman is guessing randomly.  

Choice to include the 0/1 correct guesses in the "at least extreme as 11/12 correct guesses" would be a two-sided hypothesis test, while to not include would be a one-sided hypothesis test. We may be interested in the one-sided hypothesis test to just see if the woman is "guessing correctly", but guessing 1 or fewer correct is rare enough to also be evidence against the null hypothesis of random guessing.

\begin{align*}
H_0: p=0.5 \;\;\; &&  H_0: p=0.5 \;\;\;\\
\underset{\text{are "at least as extreme"}}{\underbrace{H_A: p\not =0.5}_{\text{two-sided: 0, 1, 11 and 12}}}&& \underset{\text{are "at least as extreme"}}{\underbrace{H_A: p>0.5}_{\text{one-sided: only 11 and 12}}}
\end{align*}
Arguably, we might prefer a "one-sided" hypothesis test in this case since we are interested if the woman is "guessing *correctly*".  On the other hand, because guessing only 1 or fewer diagnoses correctly is so rare, this too is evidence against a null hypothesis of "random guessing".  E.g., if we have someone who can always guess wrongly, that's pretty weird, but we could then just always guess the opposite of their guess and always be correct.

## Fisher's Tea Experiment

Bristol was convinced she could taste the difference, and much preferred the taste of tea where the milk was poured in afterwards. Fisher didn't think that there could be a difference and proposed a hypothesis test to examine the situation.

Fisher made 8 cups of tea, 4 with milk in first and 4 with tea in first, and gave them to Dr. Bristol without her seeing how they were made and she would say if she thought the tea or the milk was poured first.  As it turned out, Dr. Bristol correctly identified if the tea or milk was poured first for all 8 of the cups. Fisher, being a skeptical statistician wanted to test if this could be happening by chance with Bristol just randomly guessing (or whether it seemed more likely that Bristol was not guessing).
